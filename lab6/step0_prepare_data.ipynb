```python
import numpy
import pandas
import sklearn.datasets
import sklearn.model_selection
```

# Instructions

- Run this script to generate data for all experiments
- This script write the data to `./data/credit.csv`

```python
data = sklearn.datasets.fetch_openml('credit-g', as_frame=True, parser='auto').frame
data.columns = data.columns.str.lower()
data.index.name = 'id'
data = data.rename(columns={"class": "label"})
data
```

```python
numpy.random.seed(0)
data = data.mask(
    numpy.random.choice(
        [False, True], 
        size=data.shape, 
        p=[0.999, 0.001]),
    numpy.nan
)
data
```

```python
data = data.dropna(subset=['label'])
data
```

```python
data_train = data.copy()
data_live = data.sample(
    n=100, 
    replace=False, 
    random_state=0, 
    axis='index'
)
```

```python
data_live_features = data_live.drop(columns=['label'])
data_live_features
```

```python
data_live_label  = data_live[['label']]
data_live_label
```

```python
export_train         = True
export_live_features = True
export_live_label    = False

if export_train         : data_train.to_csv('./data/credit.csv')
if export_live_features : data_live_features.to_csv('./data/credit.live.csv')
if export_live_label    : data_live_label.to_csv('./data/credit.live_label.csv')
```
