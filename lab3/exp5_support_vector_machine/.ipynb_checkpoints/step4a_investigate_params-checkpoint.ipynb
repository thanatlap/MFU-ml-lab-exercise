```python
import numpy
import pandas
import joblib
import seaborn

%matplotlib inline
```

# Instructions

- Read **the train data** from the CSV file and properly set the index
- Use `joblib` to load the trained model and print out the model parameters
- Again, print out thoses model parameters

```python
data_train = pandas.read_csv('./data/features.train.csv').set_index('id')
data_train
```

```python
model = joblib.load('model/model.joblib')
model
```

```python
print(f"model.coef_     : {model.coef_}")
print(f"model.intercept_: {model.intercept_}")
```

# Instructions
- Use `.predict` to produce predictions for all training samples
- Properly put the predictions into another `pandas.DataFrame`, with a column named `prediction`
- Now, **manually calculate** predictions for all training samples
- Also, properly put the predictions into `pandas.DataFrame`, with a column named `prediction`
- Verify that the predictions from `.predict` and manual calculations are equal

```python
predictions_from_sklearn = model.predict(data_train[['feature_1', 'feature_2']])
predictions_from_sklearn
```

```python
predictions_from_sklearn = pandas.DataFrame(
    data    = predictions_from_sklearn,
    index   = data_train.index,
    columns = ['prediction']
)
predictions_from_sklearn
```

```python
predictions_from_calculation = (
    data_train[['feature_1', 'feature_2']]
    .dot(model.coef_.T)
    .add(model.intercept_)
    .ge(0.)
    .astype(int)
    .rename(columns={0: 'prediction'})
)
predictions_from_calculation
```

```python
(predictions_from_calculation == predictions_from_sklearn).all()
```

```python
support_vectors = data_train.iloc[model.support_]
support_vectors
```

```python
seaborn.scatterplot(
    data_train,
    x   = 'feature_1',
    y   = 'feature_2',
    hue = 'label',
    alpha = 0.2
)

seaborn.scatterplot(
    support_vectors,
    x   = 'feature_1',
    y   = 'feature_2',
    hue = 'label',
    legend = False,
)
```
