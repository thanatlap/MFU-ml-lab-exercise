- Type the equation of the negative log likelihood function (nll) using Markdown language
- Write a function that takes a `numpy.array` of model parameters as input and returns the negative log likelihood 
- Evaluate the negative log likelihood function at $[0., 0.]$
  - Hint: it should return 519.86

````python
data_train = pandas.read_csv('./data/features.train.csv').set_index('id')
data_train
````

$\mathrm{nll_{data}(coefs)} = -1 \cdot \sum_{i} \Big\{ y_i \log \mathrm{P}(y_i=1|\mathbf{x}_i) + (1 - y_i) \log (1-\mathrm{P}(y_i=1|\mathbf{x}_i)) \Big\}$

````python
def calculate_nll(coef_):
    probabilities_from_calculations = (
        data_train[['feature_1', 'feature_2']]
        .dot(coef_)
        .apply(lambda x: 1./(1.+numpy.exp(-x)))
    )

    nll = -1. * (
        + data_train['label']        * numpy.log(probabilities_from_calculations)
        + (1. - data_train['label']) * numpy.log(1. - probabilities_from_calculations)
    ).sum()
    
    return nll
````

````python
calculate_nll([0., 0.])
````

````python
training_results = []

for i in range(1, 10):
    
    model = sklearn.linear_model.LogisticRegression(
        penalty       = 'none',
        fit_intercept = False,
        random_state  = 0,
        solver        = 'lbfgs',
        max_iter      = i,
    )

    model.fit(
        data_train[['feature_1', 'feature_2']],
        data_train['label'],
    )
    
    probabilities_from_sklearn = model.predict_proba(data_train[['feature_1', 'feature_2']])
    predictions_from_sklearn   = model.predict(data_train[['feature_1', 'feature_2']])
    
    nll = -numpy.log(
        + data_train['label']        * probabilities_from_sklearn[:, 1]
        + (1. - data_train['label']) * probabilities_from_sklearn[:, 0]
    ).sum()
    
    accuracy = (predictions_from_sklearn == data_train['label']).mean()

    training_results.append({
        'iter'     : i,
        'coef_0'   : model.coef_[0, 0],
        'coef_1'   : model.coef_[0, 1],
        'nll'      : nll,
        'accuracy' : accuracy,
    })
    
training_results = pandas.DataFrame(training_results).set_index('iter')
    
````
