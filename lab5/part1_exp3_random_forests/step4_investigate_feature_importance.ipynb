````python
import numpy
import pandas
import joblib

import seaborn
%matplotlib inline
````

# Instructions¶
- Use `joblib` to load the trained model
- Create a `pandas.DataFrame` having columns `feature_name` and `feature_importance`
    - Hint: using `feature_names_in_` and `feature_importances_` 
    - Also, sort the `feature_importance` in descending order


````python
model = joblib.load('model/model.joblib')
model
````

````python
feature_importance = pandas.DataFrame({
    'feature_name': model.feature_names_in_,
    'feature_importance': model.feature_importances_,
})

feature_importance.sort_values(
    by = 'feature_importance', 
    ascending = False,
    inplace = True
)
feature_importance
````

# Instruction 
- Use the following code snippet to plot feature importance:
```
seaborn.set(
    style = 'whitegrid', 
    rc    = {'figure.figsize': (8, 15)}
)

seaborn.barplot(
    data=feature_importance,
    x='feature_importance', 
    y='feature_name'
)
```
