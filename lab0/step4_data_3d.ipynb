````python
import numpy
import pandas
import seaborn
import matplotlib
import matplotlib.pyplot

%matplotlib inline
````

# Instructions

- In this exercise, we are going to generate grid points of size 5x5, ranging betweeb -2 to 2
- Create a numpy array of size 5, named `x_range`, whoese values are evenly spaced from -2 to 2
- Create another array of size 5, named `y_range`, whoese values are evenly spaced from -2 to 2
- Now, use `numpy.meshgrid` and `.reshape` to generate `x_coordinates`, `y_coordinates`
  - Note that these are coordinate arrays of grid points
- Then, put them into a `pandas.DataFrame` named `data` with column names `x` and `y` respectively
  - Use the default index
- Now use `seaborn.scatterplot` to plot the dataframe



````python
x_range = numpy.linspace(start=-2, stop=2, num=5)
y_range = numpy.linspace(start=-2, stop=2, num=5)
````

````python
x_coordinates, y_coordinates  = numpy.meshgrid(x_range, y_range)
````

````python
data = pandas.DataFrame({
    'x': x_coordinates.reshape(-1),
    'y': y_coordinates.reshape(-1),
})
data
````

````python
seaborn.scatterplot(
    data,
    x = 'x',
    y = 'y',
)
````
