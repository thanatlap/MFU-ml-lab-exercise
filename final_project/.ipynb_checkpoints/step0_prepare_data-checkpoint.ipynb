```python
import random
import numpy
import pandas
import sklearn.datasets
import sklearn.model_selection

pandas.set_option('display.max_rows', 4)
```

```python
data = sklearn.datasets.fetch_openml('adult', as_frame=True, parser='auto').frame
data = (
    pandas.concat([
        data[data['class'] == '<=50K']
        .sample(
            n=16_180,
            replace=False, 
            random_state=42,
        ),
        data[data['class'] == '>50K']
        .sample(
            frac=1,
            replace=False, 
            random_state=42,
        ),
    ])
    .sample(
        frac=1,
        replace=False, 
        random_state=42,
    )
    .reset_index(drop=True)
)
data
```

```python
data.columns = data.columns.str.lower()
data.index.name = 'id'

data = data.applymap(lambda s: s.lower() if isinstance(s, str) else s)
data
```

```python
country_code_mapper = {
    'united-states': 'USA',
    'cuba': 'CUB',
    'jamaica': 'JAM',
    'india': 'IND',
    'mexico': 'MEX',
    'south': 'ZAF',
    'puerto-rico': 'PRI',
    'honduras': 'HND',
    'england': 'GBR',
    'canada': 'CAN',
    'germany': 'DUE',
    'iran': 'IRN',
    'philippines': 'PHL',
    'italy': 'ITA',
    'poland': 'POL',
    'columbia': 'COL',
    'cambodia': 'KHM',
    'thailand': 'THA',
    'ecuador': 'ECU',
    'laos': 'LAO',
    'taiwan': 'TWN',
    'haiti': 'HTI',
    'portugal':'PRT',
    'dominican-republic': 'DOM',
    'el-salvador': 'SLV',
    'france': 'FRA',
    'guatemala': 'GTM',
    'china': 'CHN',
    'japan': 'JPN',
    'yugoslavia': 'YUG',
    'peru': 'PER',
    'outlying-us(guam-usvi-etc)': 'UMI',
    'scotland': 'GBR',
    'trinadad&tobago': 'TTO',
    'greece': 'GRC',
    'nicaragua': 'NIC',
    'vietnam': 'VNM',
    'hong': 'HKG',
    'ireland': 'IRL',
    'hungary': 'HUN',
    'holand-netherlands': 'NLD',
}
```

```python
data.rename(columns={'class': 'label', 'age': 'age-group'}, inplace=True)
data.insert(0, 'house-number', numpy.random.randint(100, 9999, data.shape[0]))
data.insert(0, 'social-security-number', numpy.random.randint(100_000_000, 999_999_999, data.shape[0]))
data.insert(data.columns.get_loc('native-country'), 'native-country-code', data['native-country'].map(country_code_mapper))
data
```

```python
label = (
    data['label']
    .map({
        '<=50k': 0.,
        '>50k' : 1.,
    })
    .astype(numpy.float64)
)
print(f"positive class: {label.mean()*100:.2f}%")
```

```python
numpy.random.seed(0)
data = data.mask(
    numpy.random.choice([False, True], size=data.shape, p=[0.999, 0.001]),
    numpy.nan
)
data['label'] = label
data
```

```python
data_train, data_test = sklearn.model_selection.train_test_split(
    data, 
    train_size   = 0.75, 
    shuffle      = True, 
    random_state = 42
)
```

```python
data_test_features = data_test.drop(columns=['label'])
data_test_features
```

```python
data_test_label  = data_test[['label']]
data_test_label
```

```python
data_train.to_csv('./data/high_salary.train.csv')
data_test_label.to_csv('./data/high_salary.test_label.csv')
data_test_features.to_csv('./data/high_salary.test_features.csv')
```
