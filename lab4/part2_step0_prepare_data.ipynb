```python
import numpy
import pandas
import sklearn.datasets
```

# Instructions

- Run this script to generate data for all experiments
- This script write the data to `./data/swiss_roll.csv`

```python
def rotation_matrix(theta):
    return (
        numpy.asarray([
            [1, 0, 0],
            [0, numpy.cos(theta), -numpy.sin(theta)],
            [0, numpy.sin(theta), numpy.cos(theta)],
        ])
        .dot(numpy.asarray([
            [numpy.cos(theta), 0, numpy.sin(theta)],
            [0, 1, 0],
            [-numpy.sin(theta), 0, numpy.cos(theta)],
        ]))
        .dot(numpy.asarray([
            [numpy.cos(theta), -numpy.sin(theta), 0],
            [numpy.sin(theta), numpy.cos(theta), 0],
            [0, 0, 1],
        ]))
    )
```

```python
data_points, data_colors = sklearn.datasets.make_swiss_roll(
    n_samples = 1000, 
    noise = 0.,
    random_state = 0,
)
data_points = data_points.dot(rotation_matrix(numpy.pi/6))
```

```python
data = pandas.DataFrame({
    'feature_1': data_points[:, 0],
    'feature_2': data_points[:, 1],
    'feature_3': data_points[:, 2],
    'label': (data_points.mean(axis=1) > data_points.mean()).astype(int),
})

data.index.name = 'id'
data
```

```python
data.to_csv('./data/swiss_roll.csv')
```
